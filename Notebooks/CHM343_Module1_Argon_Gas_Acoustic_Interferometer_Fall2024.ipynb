The remotely accessible acoustic interferometer is a DIY setup where a 1 inch inner diameter (ID) tube with a length of around 2 ft is capped on each end with o-ring sealed 3D printed flat end-caps that have o-ring sealed inlets for a speaker, microphone and gas inlet and outlets (Samrat Amin designed and can describe the details of the system).

For white noise data, the resulting sound data is Fourier Transformed (FT) to a frequency spectrum, where the resonant nodes can be easily observed (in audacity this can be done by going to Analyze > Plot Spectrum). ## Brief Summary of Data Analysis of Gas Acoustic Interferometry

The resonance frequencies ($ν$) are related to the the node number ($n = 1, 2, 3, …$) by the speed of sound ($c$) and the length of the acoustic interferometer tube ($L$):
<br><br>
$\nu = \left( \frac{c}{2L} \right) n$
<br><br>
Therefore, a linear fit of resonance frequencies ($ν$) and node number ($n$) gives a slope of $\left( \frac{c}{2L} \right)$ and a y-intercept of zero.  Hence, if you accurately know the speed of sound for the gas, you can determine the length of the acoustic interferometer resonance tube ($L$, which is estimated to be around 2 inches (0.6096 meters)).  This length is constant and does not change.  So, using a gas with a well known speed of sound allows the length of the tube to be accurately determined.  Once the length of the tube has been determined using one or more gases with a well known speed of sound, we can now use this length to determine the speed of sound of any gases with an unknown speed of sound.  

It is common to use Argon gas to determine the cavity (tube) length.  Argon gas sound velocity is well-known under various conditions (temperature, pressure).

In the current acoustic interferometer setup (Aug. 2024) we have the following gases connected to the computer controlled valves:

# Use the Python SciPy Library to Find Peaks in the Argon Spectral Data

The python [scipy](https://docs.scipy.org/doc/scipy/index.html) library has a [signals](https://docs.scipy.org/doc/scipy/reference/signal.html) component with a [find_peaks](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.find_peaks.html) function that makes determining the resonance frequencies more automated and convenient.

```python
# python code using SciPy library for extracting resonance 'peak' locations in acoustic interferometry white noise spectral data
Ar_peaks1 = find_peaks(Ar_Spec['Level (dB)'], rel_height=50, width= 300, distance = 200, prominence = 5)
Ar_peaks2 = find_peaks(Ar_Spec2['Level (dB)'], rel_height=50, width= 300, distance = 200, prominence = 5)
Ar_peaks3 = find_peaks(Ar_Spec3['Level (dB)'], rel_height=50, width= 300, distance = 200, prominence = 5)
Ar_peaks4 = find_peaks(Ar_Spec4['Level (dB)'], rel_height=50, width= 300, distance = 200, prominence = 5)

# Define the frequency in Hz from each of the gas spectra
Ar_Frequency1 = Ar_Spec['Frequency (Hz)']
Ar_Frequency2 = Ar_Spec2['Frequency (Hz)']
Ar_Frequency3 = Ar_Spec3['Frequency (Hz)']
Ar_Frequency4 = Ar_Spec4['Frequency (Hz)']

# Make a list of the peaks determined from the SciPy find_peak function
Ar_peak_list1 = Ar_Frequency1[Ar_peaks1[0]]
Ar_peak_list2 = Ar_Frequency2[Ar_peaks2[0]]
Ar_peak_list3 = Ar_Frequency3[Ar_peaks3[0]]
Ar_peak_list4 = Ar_Frequency4[Ar_peaks4[0]]

# NumPy array and rounding of gas peak list determined from SciPy find_peak function
Ar_peak_array1 = np.array((round(Ar_peak_list1,2)))
Ar_peak_array2 = np.array((round(Ar_peak_list2,2)))
Ar_peak_array3 = np.array((round(Ar_peak_list3,2)))
Ar_peak_array4 = np.array((round(Ar_peak_list4,2)))

# Pandas DataFrame for the frequencys of each peak determined from SciPy find_peak
Ar_peaktable1 = pd.DataFrame(Ar_peak_array1, columns=["Frequency (Hz)"])
Ar_peaktable2 = pd.DataFrame(Ar_peak_array2, columns=["Frequency (Hz)"])
Ar_peaktable3 = pd.DataFrame(Ar_peak_array3, columns=["Frequency (Hz)"])
Ar_peaktable4 = pd.DataFrame(Ar_peak_array4, columns=["Frequency (Hz)"])

# Numpy array to define Nodes 1 thru n (the number of peaks determined from SciPy find_peak)
Ar_peaktable1['Node'] = np.arange(1, Ar_peaktable1.shape[0] + 1)
Ar_peaktable2['Node'] = np.arange(1, Ar_peaktable2.shape[0] + 1)
Ar_peaktable3['Node'] = np.arange(1, Ar_peaktable3.shape[0] + 1)
Ar_peaktable4['Node'] = np.arange(1, Ar_peaktable4.shape[0] + 1)

# All peaks picked by SciPy put in a pandas dataframe table using .loc to select Node and Frequency labeled data.
Ar_peaktable1 = Ar_peaktable1.loc[:,['Node','Frequency (Hz)']]
Ar_peaktable2 = Ar_peaktable2.loc[:,['Node','Frequency (Hz)']]
Ar_peaktable3 = Ar_peaktable3.loc[:,['Node','Frequency (Hz)']]
Ar_peaktable4 = Ar_peaktable4.loc[:,['Node','Frequency (Hz)']]

# First 15 peaks for each gas dataset (the last data point from SciPy find_peak is often scewed.  Officially, we should probably do a T-test before ommitting this data)
Ar_peaktable1_15 = Ar_peaktable1.head(15)
Ar_peaktable2_15 = Ar_peaktable2.head(15)
Ar_peaktable3_15 = Ar_peaktable3.head(15)
Ar_peaktable4_15 = Ar_peaktable4.head(15)
```

### Fit and Plot the Argon Gas Resonance Frequency Data
The resonance frequencies determined from the find_peak scipy function is plotted for each gas versus the resonance or node number.  This should be linear and the slope is related to the speed of sound (and the length of the acoustic interferometer tube).  The spectral data is derived from the white noise sound datasets (a similar analysis and plot can be done for the chirp datasets).

```python
# Initialize plot layout
fig, ax = plt.subplots(figsize = (7, 5))

# Set general font size
plt.rcParams['font.size'] = '12'

# Set X and Y Axis Labels for subPlots
fig.supxlabel('Node (Resonance) Number')
fig.supylabel('Frequency (Hz)')

# Set tick font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
	label.set_fontsize(10)

# Add scatter plots
ax.scatter(Ar_peaktable1_15['Node'], Ar_peaktable1_15['Frequency (Hz)'], s=60, alpha=0.7, edgecolors="k")
ax.scatter(Ar_peaktable2_15['Node'], Ar_peaktable2_15['Frequency (Hz)'], s=60, alpha=0.7, edgecolors="k")
ax.scatter(Ar_peaktable3_15['Node'], Ar_peaktable3_15['Frequency (Hz)'], s=60, alpha=0.7, edgecolors="k")
ax.scatter(Ar_peaktable4_15['Node'], Ar_peaktable4_15['Frequency (Hz)'], s=60, alpha=0.7, edgecolors="k")

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (b) and intercept (a)
# deg=1 means linear fit (i.e. polynomial of degree 1)
b1, a1 = np.polyfit(Ar_peaktable1_15['Node'], Ar_peaktable1_15['Frequency (Hz)'], deg=1)
b2, a2 = np.polyfit(Ar_peaktable2_15['Node'], Ar_peaktable2_15['Frequency (Hz)'], deg=1)
b3, a3 = np.polyfit(Ar_peaktable3_15['Node'], Ar_peaktable3_15['Frequency (Hz)'], deg=1)
b4, a4 = np.polyfit(Ar_peaktable4_15['Node'], Ar_peaktable4_15['Frequency (Hz)'], deg=1)

# Speed of Sound for Argon Gases at 22C.
c = 320

# Determine the Acoustic Interferometer resonance tube length (L) for Gases 1 (air), 2 (Ar) and 3 (CO2) in m/s.
l1 = c/(2*b1)
l2 = c/(2*b2)
l3 = c/(2*b3)
l4 = c/(2*b4)

# Create linear sequence of points for fit line
xseq = np.linspace(0, 15, num=16)

# Plot regression line
ax.plot(xseq, a1 + b1 * xseq, alpha=0.7, color="k", lw=1.1)
ax.plot(xseq, a2 + b2 * xseq, alpha=0.7, color="k", lw=1.1)
ax.plot(xseq, a3 + b3 * xseq, alpha=0.7, color="k", lw=1.1)
ax.plot(xseq, a4 + b4 * xseq, alpha=0.7, color="k", lw=1.1)

#Annotate Plot with text
#ax.text(10, 2000, 'Ar Gas (Slope = {})'.format(round(b1,2)), fontsize=10)

ax.text(7.5, 400, 'Ar2 (c=320 m/s) - Length = {} m'.format(round(l2,4)), fontsize=8)
ax.text(7.5, 700, 'Ar3 (c=320 m/s) - Length = {} m'.format(round(l3,4)), fontsize=8)
ax.text(7.5, 100, 'Ar1 (c=320 m/s) - Length = {} m'.format(round(l1,4)), fontsize=8)
ax.text(7.5, 1000, 'Ar4 (c=320 m/s) - Length = {} m'.format(round(l4,4)), fontsize=8)

# Create a Rectangle patch
rect = patches.Rectangle((7, 0), 8, 1300, linewidth=1, edgecolor='k', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

#Show the plot in this notebook
plt.show()
```

## Error Analysis

Last component that needs to be done is to determine the statistical error associated with the data analysis (primarily the linear regression).  This is most commonly seen using the LINEST function in excel or google sheets.  The primary meteric we would like to obtain is the 95 $\%$ confidence in $\Delta L$ and/or $\Delta c$, which can then be used to report the values +/- this delta value.  Using SciPy stat a linear regression and statistical analysis is performed.  The coefficient of determination (R-squared) is determined for each linear regression fit and the 95% confidence interval on slope and intercept is calculated.

```python
# @title
# Error Analysis - White Noise Sound Data (Frequency)

# Linear Regression using SciPy Stat Library
linearregression_Ar_gas1 = stats.linregress(Ar_peaktable1_15['Node'], Ar_peaktable1_15['Frequency (Hz)'])
linearregression_Ar_gas2 = stats.linregress(Ar_peaktable2_15['Node'], Ar_peaktable2_15['Frequency (Hz)'])
linearregression_Ar_gas3 = stats.linregress(Ar_peaktable3_15['Node'], Ar_peaktable3_15['Frequency (Hz)'])
linearregression_Ar_gas4 = stats.linregress(Ar_peaktable4_15['Node'], Ar_peaktable4_15['Frequency (Hz)'])

# Print R-squared values for each of the linear regression fits
print(f"Linear Regression R-squared")
print(f"Ar Gas 1 - {linearregression_Ar_gas1.rvalue**2:.6f}")
print(f"Ar Gas 2 - {linearregression_Ar_gas2.rvalue**2:.6f}")
print(f"Ar Gas 3 - {linearregression_Ar_gas3.rvalue**2:.6f}")
print(f"Ar Gas 4 - {linearregression_Ar_gas4.rvalue**2:.6f}")
print(f" ")


# Two-sided inverse Students t-distribution, p - probability, df - degrees of freedom
tinv = lambda p, df: abs(t.ppf(p/2, df))

print(f"95% Confidence Level (+/-)")
print(f"__________________________")
ts1 = tinv(0.05, len(Ar_peaktable1_15['Node'])-2)
print(f"Ar Gas 1")
print(f"slope = {linearregression_Ar_gas1.slope:.2f} +/- {ts1*linearregression_Ar_gas1.stderr:.2f}")
print(f"y-intercept = {linearregression_Ar_gas1.intercept:.2f}"
      f" +/- {ts1*linearregression_Ar_gas1.intercept_stderr:.2f}")
print(f" ")

ts2 = tinv(0.05, len(Ar_peaktable2_15['Node'])-2)
print(f"Ar Gas 2")
print(f"slope = {linearregression_Ar_gas2.slope:.2f} +/- {ts2*linearregression_Ar_gas2.stderr:.2f}")
print(f"y-intercept = {linearregression_Ar_gas2.intercept:.2f}"
      f" +/- {ts2*linearregression_Ar_gas2.intercept_stderr:.2f}")
print(f" ")

ts3 = tinv(0.05, len(Ar_peaktable3_15['Node'])-2)
print(f"Ar Gas 3")
print(f"slope = {linearregression_Ar_gas3.slope:.2f} +/- {ts3*linearregression_Ar_gas3.stderr:.2f}")
print(f"y-intercept = {linearregression_Ar_gas3.intercept:.2f}"
      f" +/- {ts3*linearregression_Ar_gas3.intercept_stderr:.2f}")
print(f" ")

ts4 = tinv(0.05, len(Ar_peaktable4_15['Node'])-2)
print(f"Ar Gas 4")
print(f"slope = {linearregression_Ar_gas4.slope:.2f} +/- {ts4*linearregression_Ar_gas4.stderr:.2f}")
print(f"y-intercept = {linearregression_Ar_gas4.intercept:.2f}"
      f" +/- {ts4*linearregression_Ar_gas4.intercept_stderr:.2f}")
print(f" ")
```
